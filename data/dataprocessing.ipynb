```python
import pandas as pd
from datetime import datetime
import json
```

```python
data = pd.read_csv("db.csv")
```

```python
data
```

```python
ans = []
for i in data.index:
    temp = {
        "movie" : data['Original Title'][i],
        'company' : data['Company'][i],
        'imdb' : data['Rate'][i],
        'gross' : str(data['Gross Worldwide'][i]),
        'year' : str(data['Release'][i]),
        'budget' : str(data['Budget'][i])
    }
    ans.append(temp)
    # print(data['Original Title'][i])
```

```python
ans
```

```python
final = json.dumps(ans)
with open("db.json", "w") as outfile: 
    outfile.write(final) 
```
